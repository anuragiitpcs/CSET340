```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('Dog.jpeg', cv2.IMREAD_GRAYSCALE)

# Define the 3x3 averaging filter mask
mask = np.ones((3, 3), dtype=np.float32) / 9

# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, mask)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('3x3 Averaging Filter')
plt.imshow(filtered_image, cmap='gray')
plt.show()
```

```python
# Load the image
image = cv2.imread('Flower.jpeg', cv2.IMREAD_GRAYSCALE)

# Define the 3x3 weighted average filter mask
mask = np.array([[1, 2, 1],
                 [2, 4, 2],
                 [1, 2, 1]], dtype=np.float32) / 16

# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, mask)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('3x3 Weighted Average Filter')
plt.imshow(filtered_image, cmap='gray')
plt.show()
```

```python
# Load the image
image = cv2.imread('Dog.jpeg', cv2.IMREAD_GRAYSCALE)

# Define the 5x5 Gaussian filter mask
mask = np.array([[1,  4,  6,  4, 1],
                 [4, 16, 24, 16, 4],
                 [6, 24, 36, 24, 6],
                 [4, 16, 24, 16, 4],
                 [1,  4,  6,  4, 1]], dtype=np.float32) / 256

# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, mask)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('5x5 Gaussian Filter')
plt.imshow(filtered_image, cmap='gray')
plt.show()
```

```python
# Load a low-contrast image
image = cv2.imread('/content/Low Contrast 1.jpeg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Plot histograms
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(2, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')

plt.subplot(2, 2, 3)
plt.hist(image.ravel(), 256, [0, 256], color='r')
plt.title('Original Histogram')

plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), 256, [0, 256], color='b')
plt.title('Equalized Histogram')

plt.tight_layout()
plt.show()
```

```python
# Load the images
image1 = cv2.imread('Dog.jpeg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('DogLeftRotated.jpeg', cv2.IMREAD_GRAYSCALE)

# Initialize ORB detector
orb = cv2.ORB_create()

# Find keypoints and descriptors
kp1, des1 = orb.detectAndCompute(image1, None)
kp2, des2 = orb.detectAndCompute(image2, None)

# Match descriptors using BFMatcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)

# Sort matches by distance
matches = sorted(matches, key=lambda x: x.distance)

# Draw matches
matched_image = cv2.drawMatches(image1, kp1, image2, kp2, matches[:50], None, flags=2)

# Display the result
plt.figure(figsize=(10, 5))
plt.title('Image Registration using ORB')
plt.imshow(matched_image, cmap='gray')
plt.show()
```
