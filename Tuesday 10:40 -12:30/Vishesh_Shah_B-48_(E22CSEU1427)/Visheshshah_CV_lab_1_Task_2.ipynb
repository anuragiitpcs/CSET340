```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("img.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Original Image")
plt.show()
```

```python
def translate_image(image, tx, ty):
    rows, cols = image.shape[:2]
    translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
    translated = cv2.warpAffine(image, translation_matrix, (cols, rows))
    return translated
    translated = translate_image(image, 100, 30)
plt.imshow(translated)
plt.title("Translated Image")
plt.show()
```

```python
def scale_image(image, fx, fy):
    scaled = cv2.resize(image, None, fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)
    return scaled
    scaled = scale_image(image, 1.5, 1.5)
plt.imshow(scaled)
plt.title("Scaled Image")
plt.show()
```

```python
def rotate_image(image, angle):
    rows, cols = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)
    rotated = cv2.warpAffine(image, rotation_matrix, (cols, rows))
    return rotated
    rotated = rotate_image(image, 45)
plt.imshow(rotated)
plt.title("Rotated Image (45)")
plt.show()
```

```python
def reflect_image(image, axis):
    return cv2.flip(image, axis)
    flipped = reflect_image(image, 1)
plt.imshow(flipped)
plt.title("Reflected Image")
plt.show()
```

```python
def crop_image(image, x_start, y_start, width, height):
    return image[y_start:y_start + height, x_start:x_start + width]
    cropped = crop_image(image, 50, 50, 200, 200)
plt.imshow(cropped)
plt.title("Cropped Image")
plt.show()
```

```python
def shear_image(image, shx, shy):
    rows, cols = image.shape[:2]
    shear_matrix = np.float32([[1, shx, 0], [shy, 1, 0]])
    sheared = cv2.warpAffine(image, shear_matrix, (cols, rows))
    return sheared
    sheared = shear_image(image, 0.3, 0)
plt.imshow(sheared)
plt.title("Sheared Image")
plt.show()
```
